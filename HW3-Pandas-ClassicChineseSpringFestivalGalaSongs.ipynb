##**Clean the data**

```python
# Function to clean the text
def word_clear(x):
    # Delete special characters
    final=x.replace("（完）","")
    final=final.replace("（图）","")
    final=final.replace("\n","")
    final=final.replace("“","")
    final=final.replace("”","")
    # Delete half-width comma
    final=final.replace(",","")
    # Delete full-width period
    final=final.replace("。","")
    return final
```

```python
# Clean each column with the function
df["歌詞"] = df["歌詞"].map(word_clear)
df["簡介"] = df["簡介"].map(word_clear)
df["評價"] = df["評價"].map(word_clear)
df
```
