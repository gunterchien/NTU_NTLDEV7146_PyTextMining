#**Mention of Educational Stages in Taiwan in People's Daily (人民日報提及台灣各教育階段的次數)**

The purpose of this analysis was to find out to what extent People's Daily cares about the education stages in Taiwan, which include preschooling, secondary education (primary schools, junior high, and senior high), higher education (undergraduate and postgraduate), and continuing education (including adult education). The results reveal that the most frequently mentioned educational stage is higher education, followed by secondary education, preschooling, and continuing education. For secondary and higher education, it is notable that the two stages were mentioned the most in July and November of 2018, though the cause(s) of the high frequency need further investigation. As for preschooling and continuing education, the two least discussed stages of education saw a rise in March of 2018 and April of 2019 respectively, although their overall mention counts remained relatively low, generally under 15.

##**Analyze the number of education stages mentioned**

```python
# Calculate the number of words related to education
df['content_台湾'] = df['content'].str.count('台湾')
df['content_幼兒園'] = df['content'].str.count('幼兒園')
df['content_中等教育'] = df['content'].str.count('中等教育')
df['content_高等教育'] = df['content'].str.count('高等教育')
df['content_继续教育'] = df['content'].str.count('继续教育')
df[:10]
```

```python
# Select news content focusing on Taiwan (台湾)
df2 = df.loc[(df['content_台湾'] > 1)]
df2.reset_index(drop = True, inplace = True)
```

```python
# Calculate the number of education keywords mentioned
preschool = ['幼兒園', '學前教育', '小班', '中班', '大班']
secondary = ['中等教育', '小学', '初中', '高中']
higher = ['高等教育', '大学', '研究生', '硕士', '博士']
continuing = ['继续教育', '成人教育']

# Initialize empty lists for counting
num_preschool = []
num_secondary = []
num_higher = []
num_continuing = []

# Count the number of words in `preschool` (preschooling)
for n in range(len(df2)):
    count_ = 0
    for c in preschool:
        num = df.loc[n, 'content'].count(c)
        if num > 0:
            count_ += num
    num_preschool.append(count_)

# Count the number of words in `secondary` (secondary education)
for n in range(len(df2)):
    count_=0
    for c in secondary:
        num = df.loc[n, 'content'].count(c)
        if num > 0:
            count_ += num
    num_secondary.append(count_)

# Count the number of words in `higher` (higher education)
for n in range(len(df2)):
    count_=0
    for c in higher:
        num = df.loc[n, 'content'].count(c)
        if num > 0:
            count_ += num
    num_higher.append(count_)

# Count the number of words in `continuing` (continuing education)
for n in range(len(df2)):
    count_=0
    for c in continuing:
        num = df.loc[n, 'content'].count(c)
        if num > 0:
            count_ += num
    num_continuing.append(count_)

# Add columns for the counted keywords
df2["count_preschool"] = num_preschool
df2["count_secondary"] = num_secondary
df2["count_higher"] = num_higher
df2["count_continuing"] = num_continuing
df2[:10]
```

```python
# Initiate empty lists for the mentions of keywords by year and month
num_presch = []
num_second = []
num_high = []
num_cont = []

# Count the total mentions of keywords by year and month
for n in range(len(ym)):

    df3 = df2[df2.time == ym[n]]

    total_presch = df3.count_preschool.sum()
    num_presch.append(total_presch)

    total_second = df3.count_secondary.sum()
    num_second.append(total_second)

    total_high = df3.count_higher.sum()
    num_high.append(total_high)

    total_cont = df3.count_continuing.sum()
    num_cont.append(total_cont)

# Output the counts in a dataframe
data = list(zip(ym, num_presch, num_second, num_high, num_cont))
df_ym = pd.DataFrame(data, columns = ["ym", "mention_preschool", "mention_secondary", "mention_higher", "mention_continuing"])
df_ym
```

```python
# Reformat the datatime
df_ym['ym_dt'] = pd.to_datetime(df_ym['ym'], format = '%Y%m')
df_ym
```

##**Plot the number of education stages mentioned**

```python
# Prepare for a line graph
variables = ['mention_preschool','mention_secondary', 'mention_higher', 'mention_continuing']
fig, ax = plt.subplots(figsize=(16,10))

# Set the x-axis
time = df_ym.ym_dt.unique()

# Draw lines using data
for c in range(len(variables)):
  ax.plot(df_ym.ym_dt, df_ym[variables[c]], label = variables[c])

# Provide legends
ax.legend(loc = 1)         # the position of the legend
ax.set_xlabel('Time')      # x-label
ax.set_ylabel('Frequency') # y-label

# Set the title
ax.set_title("Comparisons of the number of education stages mentioned in People's Daily")
plt.show()
```
